```python
import numpy as np
import pandas as pd
```

```python
stock_data = pd.read_csv(open("data.csv"))
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(init='random',n_clusters=8)
```

```python
stock_data2 = stock_data.copy()
```

```python
stock_data2 = stock_data2.drop("StockName",axis=1)
```

```python
kmeans.fit(stock_data2)
```

### Question 2

trying to decrease SSE with k constant

n_init : increasing it will decrease SSE 

```python
kmeans_1 = KMeans(init='random',n_init= 100,n_clusters=8)
kmeans_1.fit(stock_data2)
kmeans_1.inertia_
```

this is what we observe to a certain extent (till 3000) 1560 seems to be the best value we can get

```python
kmeans_1 = KMeans(init='random',n_init= 100,n_clusters=8)
kmeans_1.fit(stock_data2)
kmeans_1.inertia_
```
